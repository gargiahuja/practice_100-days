```python

from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
print(x)
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red')
```

```python
y_pred = ((78.35 * x) + 100).reshape(4)
```

```python
#applying gradient descent
m=78.35
b=100
lrr=0.1
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
step_size=lrr*loss_slope
b=b-step_size
```

```python
y_pred1 = ((78.35 * x) + b).reshape(4)
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='blue',label='b=0')
plt.plot(x,y_pred1,color='green',label='b = {}'.format(b))
plt.legend()
plt.show()
```

```python
loss_slope=-2*np.sum(y-m*x.ravel()-b)
step_size=loss_slope*lrr
b=b-step_size
y_pred3=((78.35*x)+b).reshape(4)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='blue',label='b=100')
plt.plot(x,y_pred1,color='yellow',label='b = {}'.format(b))
plt.plot(x,lr.predict(x),color='green',label='OLS')
plt.plot(x,y_pred3,color="#000000",label='b = {}'.format(b))
plt.legend()
plt.show()
```
