```python
df=pd.read_csv(r"F:\ML\100 days ml\data\placement2.csv")
```

```python
df.shape
```

```python
min=df['cgpa'].mean()-3*df['cgpa'].std()
max=df['cgpa'].mean()+3*df['cgpa'].std()
```

```python
df['cgpa']=np.where(df['cgpa']<min,min,np.where(df['cgpa']>max,max,df['cgpa']))
```
