```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(x,y)
```

```python
b=100
m=-100
all_m=[]
all_b=[]
lr=0.001
```

```python
for i in range(30):
    slope=0
    cost=0
    for j in range(x.shape[0]):
        slope_m=-2*(y[j]-(m*x[j]+b))*x[j]
        slope_b=-2*(y[j]-(m*x[j]+b))
        cost=cost+(y[j]-m*x[j]-b)**2
    
    m=m-(lr*slope_m)
    b=b-(lr*slope_b)
    all_m.append(m)
    all_b.append(b)
    y_pred=m*x+b
    plt.scatter(x,y,color='red')
    plt.plot(x,y_pred,color='blue')
    plt.show()
```

```python
for i in range(30):
    slope_m = 0
    slope_b = 0
    cost = 0

    # Loop over each data point
    for j in range(len(x)):
        y_pred_j = m * x[j] + b
        error = y[j] - y_pred_j
        slope_m += -2 * error * x[j]
        slope_b += -2 * error
        cost += error**2

    # Update parameters using average gradient
    m = m - lr * (slope_m)
    b = b - lr * (slope_b)

    all_m.append(m)
    all_b.append(b)

    # Plotting the fit line at this step
    y_pred = m * x + b
    plt.plot(x, y_pred, color='blue')
    plt.title(f"Iteration {i+1}")
plt.scatter(x, y, color='red')
plt.show()
```
